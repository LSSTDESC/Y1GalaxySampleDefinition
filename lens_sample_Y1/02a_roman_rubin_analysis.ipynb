## Studying SNR and mag limited cuts

```python
catal_snr_i = catalog[(catalog['mag_i_lsst']<23.7) & (catalog['mag_err_i_lsst']<2.5*np.log10(1+0.1))]
catal_i = catalog[catalog['mag_i_lsst']<23.7]
catal_snr = catalog[catalog["mag_err_i_lsst"]< 2.5*np.log10(1+0.1)]
```

```python
z_bins = np.arange(0, 3, 0.1)
fractions = []
fractions_snr = []
fractions_i = []

for i in range(len(z_bins) - 1):
    z_min = z_bins[i]
    z_max = z_bins[i + 1]
    
    catal_snr_i_cut = catal_snr_i[(catal_snr_i['redshift'] > z_min) & (catal_snr_i['redshift'] <= z_max)]
    catal_i_cut = catal_i[(catal_i['redshift'] > z_min) & (catal_i['redshift'] <= z_max)]
    catal_snr_cut = catal_snr[(catal_snr['redshift'] > z_min) & (catal_snr['redshift'] <= z_max)]
    catalog_cut = catalog[(catalog['redshift'] > z_min) & (catalog['redshift'] <= z_max)]
    
    fractions_snr.append(len(catal_snr_cut)/len(catalog_cut))
    fractions_i.append(len(catal_i_cut)/len(catalog_cut))
    
    fraction = len(catal_snr_i_cut) / len(catal_i_cut) if len(catal_i_cut) > 0 else 0
    fractions.append(fraction)
    
fractions = np.array(fractions)
fractions_i = np.array(fractions_i)
fractions_snr = np.array(fractions_snr)
```

```python
bin_centers = (z_bins[:-1] + z_bins[1:]) / 2

plt.figure(figsize=(12, 8))
plt.plot(bin_centers, fractions, marker='o', linestyle='-', color='b', label='i band + SNR cut / i band cut')
plt.plot(bin_centers, fractions_i, marker='o', linestyle='-', color='green', label='i band cut / total')
plt.plot(bin_centers, fractions_snr, marker='o', linestyle='-', color='red', label='SNR cut / total')
plt.xlabel('Redshift', fontsize=18)
plt.ylabel('Fraction', fontsize=18)
plt.grid(True)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.legend(fontsize=18)
plt.title('Fraction of galaxies after cut in SNR and in i band', fontsize=22)
plt.show()
```
